```python
from os import name
import xml.etree.ElementTree as et
import pandas as pd
from bs4 import BeautifulSoup
from lxml import html
from urllib.parse import urlencode, quote_plus, unquote


#bs4 사용하여 item 태그 분리

xml_obj = BeautifulSoup(content,'lxml-xml')
rows = xml_obj.findAll('item')
print(rows)
```

```python
# 각 행의 컬럼, 이름, 값을 가지는 리스트 만들기
row_list = [] # 행값
name_list = [] # 열이름값
value_list = [] #데이터값

# xml 안의 데이터 수집
for i in range(0, len(rows)):
    columns = rows[i].find_all()
    #첫째 행 데이터 수집
    for j in range(0,len(columns)):
        if i ==0:
            # 컬럼 이름 값 저장
            name_list.append(columns[j].name)
        # 컬럼의 각 데이터 값 저장
        value_list.append(columns[j].text)
    # 각 행의 value값 전체 저장
    row_list.append(value_list)
    # 데이터 리스트 값 초기화
    value_list=[]
```

```python
corona_df = pd.DataFrame(row_list, columns=name_list)
```

```python
from os import name
import xml.etree.ElementTree as et
import pandas as pd
from bs4 import BeautifulSoup
import requests
import time
import os 

final_df = pd.DataFrame()
years = list(range(2010, 2023))  # 2010년부터 2022년까지
sick_cd_list = corona_df['sickCd'].to_list()  # 여기에 반복할 sickCd 값을 추가
url = "http://apis.data.go.kr/B551182/diseaseInfoService/getDissByGenderAgeStats"
service_key =  'f9QglMNSs8MZXJ2yt4grBrAPXOM/7aShpJfjjVNST5zz1bsrmjsLWohfJeHpaU3fQkIGJXyB/m4slxWqcNkmEQ==',
# 각 연도와 `sickCd`에 대해 반복

for sickCd in sick_cd_list:
    # 데이터프레임 초기화
    final_df = pd.DataFrame()

    for year in years:
        # 요청 실패 시 재시도를 위한 제한
        max_retries = 5  # 최대 재시도 횟수
        retry_count = 0  # 현재 재시도 횟수
        success = False  # 요청 성공 여부
        
        while retry_count < max_retries and not success:
            try:
                # 파라미터 구성
                params = {
                    'serviceKey': service_key,
                    'numOfRows': 100,
                    'pageNo': 1,
                    'year': year,
                    'sickCd': sickCd,
                    'sickType': 1,
                    'medTp': 2
                }

                # HTTP 요청 보내기
                response = requests.get(url, params=params, verify=False)  # SSL 검증 비활성화 (주의)
                
                if response.status_code != 200:
                    raise Exception(f"Failed to get data: {response.status_code}")  # 오류 발생 시 예외
                
                # XML 데이터 파싱
                soup = BeautifulSoup(response.text, 'xml')
                items = soup.find_all('item')  # 'item' 추출
                
                # 데이터프레임에 추가
                row_list = []
                for item in items:
                    row = {tag.name: tag.text for tag in item.find_all()}
                    row['year'] = year  # 연도 추가
                    row_list.append(row)
                
                # 데이터프레임에 추가
                temp_df = pd.DataFrame(row_list)
                final_df = pd.concat([final_df, temp_df], ignore_index=True)

                success = True  # 요청 성공 시 플래그 설정

            except Exception as e:
                retry_count += 1  # 재시도 횟수 증가
                print(f"Error: {e}. Retrying ({retry_count}/{max_retries})...")
                time.sleep(2 * retry_count)  # 지연 시간 증가
                
        # 최대 재시도 후에도 실패한 경우
        if not success:
            print(f"Skipping year {year} for sickCd {sickCd} after {max_retries} retries.")
            continue  # 실패한 경우 건너뛰기
        
        # 데이터프레임에 데이터 추가 후 CSV 파일로 저장
        file_name = f"sickCd_{sickCd}.csv"  # `sickCd` 별로 파일 저장
        if os.path.exists(file_name):
            final_df.to_csv(file_name, mode='a', index=False, header=False)  # 기존 파일에 추가
        else:
            final_df.to_csv(file_name, mode='w', index=False, header=True)  # 새로운 파일에 저장
        
        # 요청 간 지연 추가 (서버 과부하 방지)
        time.sleep(1)  # 지연 시간 추가
        
    # 메모리 확보를 위해 데이터프레임 초기화
    final_df = pd.DataFrame()  # 데이터프레임 비우기
```
